```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import requests
import openpyxl
import io
from PIL import Image
import time

# Path to the chrome webdriver on my PC
PATH = "C:\\Users\\niloo\\OneDrive\\Desktop\\ChromeDriver\\chromedriver.exe"

# Create a new Excel workbook and sheet
excel = openpyxl.Workbook()
sheet = excel.active
sheet.title = 'Nike Scraped Data'
sheet.append(['Title', 'Category', 'Price', 'Image Path', 'Image URL'])

wd = webdriver.Chrome(PATH)

def scroll_down(wd):
    wd.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(delay)

def get_data_from_Nike(wd, delay, sheet):

    url = "https://www.nike.com/ca/w/mens-tops-t-shirts-9om13znik1"
    wd.get(url)

    try:
        # Extract the maximum number of images from the webpage
        max_images = int(wd.find_elements(By.CLASS_NAME, "wall-header__item_count")[0].text[1:-1])
    except:
        # Handle case when the maximum number of images is not found in the specified class name
        # since it varies based on the size of opened chrome window
        max_images = int(wd.find_elements(By.CLASS_NAME, "subheading__result-count")[0].text.split(' ')[0])

    image_urls = []

    while len(image_urls) < max_images:
        scroll_down(wd)

        thumbnails = wd.find_elements(By.CLASS_NAME, "css-1fxh5tw")
        prices = wd.find_elements(By.CLASS_NAME, "css-11s12ax")
        titles = wd.find_elements(By.CLASS_NAME, "product-card__title")
        subtitles = wd.find_elements(By.CLASS_NAME, "product-card__subtitle")

        for price, thumbnail, title, subtitle in zip(prices, thumbnails, titles, subtitles):
            if thumbnail.get_attribute('src') in image_urls:
                continue

            if thumbnail.get_attribute('src') and 'http' in thumbnail.get_attribute('src'):
                # Append data to the sheet
                image_urls.append(thumbnail.get_attribute('src'))
                print('\n\n' + f"Found {len(image_urls)}")
                print(price.text)
                print(title.text)
                print(subtitle.text)
                sheet.append([subtitle.text, title.text, price.text, 'images NIKE/' + str(len(image_urls) - 1) + ".jpg", thumbnail.get_attribute('src')])

            if len(image_urls) >= max_images:
                break

    return image_urls

def download_image(download_path, url, file_name):
    try:
        # Download the image content from the URL
        image_content = requests.get(url).content
        image_file = io.BytesIO(image_content)
        image = Image.open(image_file)
        file_path = download_path + file_name

        # Save the image to the specified file path
        with open(file_path, "wb") as f:
            image.save(f, "JPEG")

    except Exception as e:
        print('FAILED -', e)

delay = 1
# Get the text data & image URLs from Nike website
urls = get_data_from_Nike(wd, delay, sheet)

# Download the images and save them to the local directory
for i, url in enumerate(urls):
    download_image("C:/Users/niloo/OneDrive/Desktop/ChromeDriver/images NIKE/", url, str(i) + ".jpg")

# Save the Excel workbook
excel.save('Nike Scraped Data.xlsx')

# Quit the webdriver
wd.quit()
```
